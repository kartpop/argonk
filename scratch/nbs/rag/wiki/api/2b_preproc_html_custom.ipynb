```python
from pathlib import Path
from haystack.components.converters import TextFileToDocument

converter = TextFileToDocument()

docs = converter.run(sources=[Path("dinosaur-page.html")])
print(docs)
```

```python
print(docs["documents"][0].content)
```

```python
html_content = docs["documents"][0].content

from bs4 import BeautifulSoup
import re

soup = BeautifulSoup(html_content, 'html.parser')

# Initialize an empty list to store the chunks
chunks = []

def clean_text(text):
    # Replace newlines and non-breaking spaces with regular spaces
    cleaned_text = text.replace('\n', ' ').replace('\xa0', ' ')
    # Replace multiple spaces with a single space
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    # Replace spaces within numbers with no space
    cleaned_text = re.sub(r'(\d)\s+(\d)', r'\1\2', cleaned_text)
    # Optionally, you can also strip leading/trailing whitespace
    cleaned_text = cleaned_text.strip()
    return cleaned_text

html_soup = soup.body or soup
for tag in html_soup.find_all(recursive=False):
    if tag.name == 'p':
        # Extract plain text from <p> tags with spaces between tags
        chunks.append(clean_text(tag.get_text(separator=' ')))
    elif tag.name == 'link':
        continue
    else:
        # Append the HTML content of other tags
        chunks.append(str(tag))

# Return the list of strings
print(chunks)
```
