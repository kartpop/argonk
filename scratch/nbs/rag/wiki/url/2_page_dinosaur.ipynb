### Parse wiki html

```python
import json

# Read the JSON file
with open('dinosaur-page.json') as file:
    data = json.load(file)

# Drill down into ["parse"]["text"]["*"]
page_html = data["parse"]["text"]["*"]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page_html, 'html.parser')
headings = soup.find_all('div')

for heading in headings:
    print(heading.text.strip())
```
